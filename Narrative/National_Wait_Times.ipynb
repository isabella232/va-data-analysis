```python
import pandas as pd
import re
import numpy as np
from scipy import stats

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from dateutil import parser
from datetime import datetime

from math import ceil
```

# The Data

```python
national_complete = pd.read_csv("../Wait_Time/Completed_Cleaned/National_Complete.csv")
national_pending = pd.read_csv("../Wait_Time/Pending_Clean/National_Pending.csv")

#used to strip time from Pending_Time
def get_date(test):
    test = test.strip('National: ')
    return parser.parse(test)

national_complete["date"] = national_complete["Complete_Time"].apply(get_date)
national_pending["date"] = national_pending["Pending_Time"].apply(get_date)
national_complete.sort_values(by="date")
national_pending.sort_values(by="date")

national_pending = national_pending.drop(20)
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(15,4))

national_complete[["date","PC_Wait_Complete_National"]].set_index("date").plot(ax=ax2)
national_complete[["date","MH_Wait_Complete_National"]].set_index("date").plot(ax=ax2)
national_complete[["date","SC_Wait_Complete_National"]].set_index("date").plot(ax=ax2)

national_pending[["date","PC_Wait_Pending_National"]].set_index("date").plot(ax=ax1)
national_pending[["date","MH_Wait_Pending_National"]].set_index("date").plot(ax=ax1)
national_pending[["date","SC_Wait_Pending_National"]].set_index("date").plot(ax=ax1)

ax1.grid(True)
ax2.grid(True)
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True, figsize=(8,8))

ax2.set_title("Completed Appointments")
national_complete[["date","PC_Wait_Complete_National"]].set_index("date").plot(ax=ax2)
national_complete[["date","MH_Wait_Complete_National"]].set_index("date").plot(ax=ax2)
national_complete[["date","SC_Wait_Complete_National"]].set_index("date").plot(ax=ax2)

ax1.set_title("Pending Appointments")
national_pending[["date","PC_Wait_Pending_National"]].set_index("date").plot(ax=ax1)
national_pending[["date","MH_Wait_Pending_National"]].set_index("date").plot(ax=ax1)
national_pending[["date","SC_Wait_Pending_National"]].set_index("date").plot(ax=ax1)

ax1.grid(True)
ax2.grid(True)

ax2.legend_.remove()
```
