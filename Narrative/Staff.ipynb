```python
import pandas as pd

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

# The Data

```python
df = pd.read_csv("../Staff_Size/Onboard_By_statyion_By_FY.csv").head(1).drop("ORGANIZATION",1)
df
```

# How has staffing changed?

```python
ax = df.transpose().plot(kind="line", color="teal")
ax.set_ylim(0,320000)
ax.set_title("VA Staffing Levels By Fiscal Year")
```

## It looks like there's a pretty steady increase...but what happens when we subtract the choice hires? We should see a steady increase without choice hires since that should have been on top of usual hires

```python
choice_hires = pd.read_csv("../Hiring/Choice VA Hiring - #4 - Choice Act Hires -Fac List.csv")
```

```python
increase_dict = {}

increase_dict["11-13"] = (df["MAY-FY13"] - df["MAY-FY11"])/df["MAY-FY11"]
increase_dict["12-14"] = (df["MAY-FY14"] - df["MAY-FY12"])/df["MAY-FY12"]
increase_dict["13-15"] = (df["MAY-FY15"] - df["MAY-FY13"])/df["MAY-FY13"]
increase_dict["14-16"] = (df["MAY-FY16"] - df["MAY-FY14"])/df["MAY-FY14"]
increase_dict["14-16_no_choice_hires"] = (df["MAY-FY16"] - df["MAY-FY14"] - choice_hires["NbrEmps"].sum())/df["MAY-FY14"]

increase = pd.DataFrame.from_dict(increase_dict)
increase
```

```python
ax = increase.transpose().plot(kind="bar",color = "teal", edgecolor="none")
ax.set_title("Increase In VA Staffing For Two Year Periods")
ax[0].set_color('r')
#ax.grid(True)
```

### Staffing level increases slowed down slightly between 2014 and 2016, even before you add in the choice hires. Without the choice hires, it's even worse.
